```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
# Series,類似一維陣列，含有值的序列
obj=pd.Series([1,2,5,698])
print(obj)
print(obj.values)
print(obj.index)

#更改index
obj=pd.Series([1,2,5,698],index=['a','b','c','d'])
# 或是obj.index=['a','b','c','d']
print(obj)
print(obj.index)




```

```python
# 透過index取值
print(obj['a'])
print(obj[['a','b','c']])

# 可配合數學含式使用
print(obj > 10)
print(obj[obj > 10])
print(obj * 2)

# Series中index,value的概念和dict相似
info={'taiwan':100000,'china':256384,'japan':85642}
obj2=pd.Series(info)
print(obj2)

# 可分別傳入value index
country=['korea','taiwan','china','japan','america',]
obj3=pd.Series(info,index=country)
print(obj3)

```

```python
# check null value
print(pd.isnull(obj3))
print(pd.isna(obj3))

# 運算時，Series會自動和index對齊
print(obj2+obj3)

# 類似database 命名/join的功能
obj3.name='population'
obj3.index.name='country'
print('-----------------')
print(obj3)
```

```python
# Dataframe，含有資料的方形資料表，每個欄位都可以是不同型態
# 建立dataframe方法:等長list物件的dict或numpy物件建立
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada','Nevada'], 
                  'year': [2000, 2001, 2002, 2001, 2002, 2003], 
                  'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data)
frame
```

```python
# dataframe常用方法
print(frame.head())

#重新排列column順序
print(pd.DataFrame(data,columns=['year','state','pop']))

# 傳入的column若不存在，以遺失值顯示
frame2=pd.DataFrame(data,columns=['year','state','pop','dept'])
print(frame2)

# 取得dataframe中的欄位，型態是Series
frame2['state']
# OR frame2.year
```

```python
# 給值可以修改欄位值
frame2['dept']=10
print(frame2)

# 用數學函式新增欄位
frame2['eastern']=frame2.state=='Ohio'
print(frame2)

# 刪除欄位
# del frame2['income']

# 如果用list/array指定給某一欄，數量要和dataframe中的相同，但是用Series指定給欄時，index會自動對齊!!!!!!!
# 注意!!!!任何對Series進行修改都會影響原本dataframe中的資料，可用copy()備份
val=pd.Series([12561,475363,65624],index=[1,2,3])
frame2['income']=val
print(frame2)

```

```python
# dict傳入dataframe
# 巢狀dict，會把外層key視為欄index(taiwan)，內層key視為列index(2001)
pop={'taiwan':{2001:2.4,2002:2.9},'japan':{2001:3.6,2002:4.5,2003:5.3}}
frame3=pd.DataFrame(pop)
print(frame3)

# 轉置dataframe(欄列交換)
print(frame3.T)

#　設定index,column的name屬性
frame3.index.name='year'
frame3.columns.name='country'
print(frame3)


```

```python
# index,index不可直接更改=> obj.index[1]='d'
# pandas的index可以重複!!
obj=pd.Series(range(3),index=['a','b','c'])
print(frame3.columns)
print(frame3.index)

print(frame3)
print('taiwan' in frame3.columns)
print(2001 in frame3.index)

# reindex() method=ffill可以填值
obj=pd.Series(['blue','yellow','orange','purple'],index=[0,2,5,9])
print(obj.reindex(range(10),method='ffill'))




```

```python
# 對column或row做reindex
obj=pd.DataFrame(np.arange(9).reshape(3,3),index=['a','c','d'],columns=['ohio','texas','washinton'])
print(obj)
print(obj.reindex(['a','b','c','d']))
print(obj.reindex(columns=['ohio','texas','utah','washinton']))
```

```python
# Series刪除特定軸
obj=pd.Series(np.arange(5),index=['a','b','c','d','e'])
print(obj)
print(obj.drop('c'))
print(obj.drop(['a','e']))

# dataframe刪除特定軸
obj=pd.DataFrame(np.arange(16).reshape(4,4),index=['ohio','texas','utah','washinton'],columns=['a','b','c','d'])
print(obj)
print(obj.drop(['ohio','texas']))
print(obj.drop('b',axis=1)) #or axis='columns'
# drop若參數inplace=True,則會更改原資料
```

```python
# series的index做切片
obj=pd.Series(np.arange(4),index=['a','b','c','d'])
print(obj['a':'c'])
obj['a','b']=10
print(obj)

```

```python
# dataframe的index做切片
data = pd.DataFrame(np.arange(16).reshape(4,4),index=['ohio','texas','utah','new york'],columns=['one','two','three','four'])
print(data)
print(data[['two','four']])
print([data>5])

# data[data<5]=0
print(data)
```

```python
#loc(軸標籤),iloc(整數)選取列和欄
print(data)
print(data.loc['ohio',['two','three']])
print(data.iloc[3]) # 第4行
print(data.iloc[2,[3,0,1]]) # 第3行的4,1,2欄位
print(data.iloc[[1,2],[0,1]]) #第2,3行的1,2欄位

# 切片
print(data.loc[:'utah','three'] )
print(data.iloc[:3,:1])
print(data.iloc[:,:3][data.three>5]) # 每一列的前三欄且欄位值>5




```

```python
# Series和dataframe皆可做算術運算和資料對齊
```

```python
# 填值
df1=pd.DataFrame(np.arange(12).reshape(3,4),columns=list('abcd'))
df2=pd.DataFrame(np.arange(20).reshape(4,5),columns=list('abcde'))
print(df1)
print(df2)
print(df1+df2)

# 使用add(),參數fill_value
print(df1.add(df2,fill_value=0)) # 把df1沒有的欄位用df2的替代

# reindex()也可以填不同的值
print(df1.reindex(columns=df2.columns,fill_value=0))
```

```python
# numpy陣列的broadcasting
arr=np.arange(12).reshape(4,3)
print(arr)
print(arr-arr[0]) # 減法對每一列都做一次，被稱為廣播
 
```

```python
#　dataframec和series之間的動作和廣播雷同
frame=pd.DataFrame(np.arange(12).reshape(4,3),columns=list('bde'),index=['utah','ohio','texas','oregon'])
print(frame)

series=frame.iloc[0]
print(series)
print(frame-series)

series2=pd.Series(range(3),index=list('bef')) # index在dataframe中匹配不到 or series中找不到，以聯集結果reindex()
print(frame+series2)
```

```python
# 用index匹配,然後對欄做廣播
series3=frame['d']
print(series3)
frame.sub(series3,axis="index") #sub()減法,傳入的軸為index


```

```python
# 全域函式
frame=pd.DataFrame(np.random.randn(12).reshape(4,3),columns=list('bde'),index=['utah','ohio','texas','oregon'])
print(frame)
print(np.abs(frame))

# 逐欄或逐列套用 apply(),可傳lambda
f=lambda x: x.max()-x.min() 
print(frame.apply(f)) # 三行的最大差值
print(frame.apply(f,axis='columns')) # 三列的最大差值

```

```python
# applymap()可傳function
def f(x):
    return pd.Series([x.min(),x.max()],index=['min','max'])
print(frame.apply(f))

formation=lambda x: '%.2f' % x
print(frame.applymap(formation))

print(frame['e'].map(formation)) # series物件才可使用
    
```

```python
# index排序 series
obj=pd.Series(range(4),index=list('ldhj'))
print(obj)
print(obj.sort_index())



```

```python
# index排序 dataframe
frame=pd.DataFrame(np.arange(8).reshape(2,4),index=['three','one'],columns=['d','a','b','c'])
print(frame)
print(frame.sort_index())
print(frame.sort_index(axis=1))
print(frame.sort_index(axis=1,ascending=False))
```

```python
# value排序 dataframe
frame=pd.DataFrame({'b':[4,7,-3,2],'a':[0,1,0,1]})
print(frame)
print(frame.sort_values(by='b'))
print(frame.sort_values(by=list('ba')))
print(frame.sort_values(by=list('ab')))
```

```python
# value排序 排序 series 
obj2=pd.Series([4,5,2,-6])
print(obj2)
print(obj2.sort_values())
```

```python
# rank()排名 series
ser = pd.Series([5, 0, 3, 8, 4, 5])
print(ser)
print(ser.rank()) # 按照index順序，比較值的大小，最小的為1，最大為n,4.5為平均排名
print(ser.rank(method='first')) # 當出現相通資料時以誰先出現為原則
print(ser.rank(ascending=False,method='max')) # 降冪排序，相同資料出現時以最大排名0
```

```python
# rank()排名 dataframe
frame=pd.DataFrame({'a':[0,1,0,1],'b':[4.3,7,-3,2],'c':[-2,5,8,-2.5]})
print(frame)
print(frame.rank()) #　值相同：(1+2)/2=1.5　(3+4)/2=3.5
print(frame.rank(axis='columns')) # 對列做rank()
```

```python
# series的index可以重複
obj=pd.Series(range(5),index=list('aabbc'))
print(obj)
print(obj.index.is_unique) # 是否沒有重複index
print(obj['a']) # 重複index,回傳series
print(obj['c']) # 回傳常數
```
