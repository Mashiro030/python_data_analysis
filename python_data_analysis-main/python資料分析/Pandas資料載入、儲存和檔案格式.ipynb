```python
#　解析xml(1)
from lxml import objectify
path='./data/ch6/mta_perf/Performance_MNR.xml'
parsed=objectify.parse(open(path))
root=parsed.getroot() 
root
```

```python
# root.INDICATOR,製作產生器
data=[]
skip_field=['PARENT_SEQ','INDICATOR_SEQ','DESIRED_CHANGE','DESIRED_CHANGE','DECIMAL_PLACES']

for element in root.INDICATOR: # 抓取在INDICATOR內所有TAG的內容
    element_dict={}
    for child in element.getchildren():　# 如果TAG在skip_field內，則跳過
        if child.tag in skip_field:
            continue
        element_dict[child.tag]=child.pyval　# 建立dict
    data.append(element_dict) # 最後append到字串中

# list轉dataframe
pd.DataFrame(data)

    
```

```python
# pandas內建２種二進位檔案格式 HDF5/MessagePack
df=pd.DataFrame({'a':np.random.randn(100)})
store=pd.HDFStore('./data/ch6/storage_type.h5')
store['obj']=df # 設定值
store['obj_col']=df['a'] # 設定index
store
```

```python
# HDFStore支援fixed,table兩種儲存方法,其中table支援查詢語法
store.put('obj2',df,format='table')
store.select('obj2',where=['index > 10 and index < 15'])

#put等同於store['obj2']=df
```
